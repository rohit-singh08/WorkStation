```python
#mean by year
df1 = df.groupby("year").mean()
df1
```

```python
df1 = df1.reset_index()
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x=df1["year"], y=df1["average_price"])
plt.xlabel('Year(1995-2021)')
plt.ylabel('Average Price of Honey')
plt.plot()
plt.savefig('Average Price of Honey vs Year')
```

```python
df1[df1["production"] == df1["production"].max()]
```

```python
df1["year"][df1["production"] == df1["production"].max()]
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x=df1["year"], y=df1["production"])
plt.xlabel('Year(1995-2021)')
plt.ylabel('Total Honey Production(Pounds)')
plt.plot()
plt.savefig('Total Honey Production vs Year.png')
```

```python
plt.figure(figsize=(10,10))
plt.pie(df1["production"], labels = df1['year'], autopct = "1%.2f%%")
plt.plot()
plt.savefig('% Distribution per Year')
```

```python
plt.plot(df1['year'], df1["production"])
plt.show()
plt.savefig('Total Honey Production vs Year')
```

```python
df2 = df.groupby(['year']).mean()
```

```python
df2 = df2.reset_index()
```

```python
df2.sort_values("value_of_production" , ascending = False)
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x=df2["year"], y=df2["value_of_production"])
plt.xlabel('Year(1995-2021)')
plt.ylabel('Value of production(Dollar)')
plt.plot()
plt.savefig('Value of production(Dollar) vs Year')
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x=df2["year"], y=df2["value_of_production"])
plt.xlabel('Year(1995-2021)')
plt.ylabel('Value of production(Dollar)')
plt.plot()
plt.savefig('Value of production(Dollar) vs Year')
```

```python
plt.figure(figsize=(10,10))
plt.pie(df2["value_of_production"], labels = df2['year'], autopct = "1%.1f%%")
plt.plot()
plt.savefig('Pie Graph of Value of Production')
```
